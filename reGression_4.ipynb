```python
# Vous savez maintenant comment fontcionne la rérgression linéaire simple et comment l'implémenter en Python.
# maintenant nous allons apprendre la rgréssion liinéaire multiple. Vous pouvez voire le représentation graphique que la régression linéaire simple sur 
# la gaucge de l'écran et sur la régression linéaire multiple sur le côté de droite de votre écran
# La différence est uniquement due au fait que la régression linéaire multiple a plus d'une entrée ou variable indépendante. Cela signifie que la sortie dépend de # chaque entrée. Par conséquent il y aura une corrélation linéaire unique entre chaque entrée et la sortie. Ces corrélations sont représentés par quatre lignes 
# différentes sur le graphique à droite de l'écran. L'illustration mathématique de la régression linéaire multiple est y= A+B1x1+B2x2+B3x3+B4x4 A est l'ordonnée
# à l'origine et B1 le pente de la ligne qui représente la relation entre la production de la première variabke indépendante B1, indique que si nous augmentons, 
# la vlaeur de X1 d'une unité B nous indique de combien la valeur de la production sera affectée.
# Quand à B2, B3, et B4, ils fonctionnent tous de manière similaire. 
# EXPORTATION DE DONNEES
# Regression linéaire multiple en Python
# Pour la mise en ooeuvre de la regression linéare multiple, nous allons utiliser un fichier csv contenant 5 collonnes et 50 lignes. Dans ces données nous 
# avons les quatre variables indépnedantes appelées R & D Spend,Administration, Marketing Spend and State et il y a une variable dépendante,  profit. Notre tâche
# ici est d'entraîner un modèle dâpprentissage automatique machine learning avec ces données, afin de comprendre la corrélation entre chacune des quatre variables # indépendantes et de prédire un bénéfice pour une autre nouvelle entreprise lorsque ces quatre variables indépendantes sont fournies.
# alors dans Juyter Notebook 



# Displaying the first few rows of the dataset (optional)

import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd

# Reading the dataset
dataset = pd.read_csv("C:/Users/micha/Downloads/DataAnl/50_Startups.csv")
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1].values

# Displaying the first few rows of the dataset (optional)
print(dataset.head())


#always create a notebook file 
# take care of spellings to avoid syntax errors


```

```python
# Implementation en python: Coddage de données catégorielles
# Si les valeurs sont des nombres dans l'ensemble de données, nous pouvons facilement appliquer des calculs mathématiques sur l'ensemble
# et créer dess modèles de prédiction. Dans cet ensemble de données nous rencontrons une variable non numérique non numérique comme la variqable State
# Ces données sont appelées données catégorielles.
# Nous devons convertir ces valeurs catégorielles en valeurs numériques d'une manière ou d'une autre.
# Nous codons ces données catégorielles en utilisant une autre bibliothèque importante appelée SK Learn

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer (transformers=[('encoder' ,OneHotEncoder(), [3])], remainder = 'passthrough')
X = np.array(ct.fit_transform(X))
# Dans ce morceau de code, nous avons emporté le colon transformé et le OneHotEncoder pour transformer notre collonne State indexée numér0 trois
# en données numériques. OneHotEncoder est un processus d'encodage ou chaque étiquette est repésentée par un tableau d'entiers plûtot car par un seul
# entier. En outre vous pouvez voire que nous avons utilisé la classe clonnes transformées. Cela permet de transformer séparément une colonne 
# particulière du Data Frame
# Le constructeur de colonnes transformé prend un certain nombre d'arguments, mais seuls deux nous intéressent. Le premier argument est un tableau
# appelé transformers, qui est une liste de tuples, Le tableau contient les éléments suivant dans le même ordre non? Un nom pour le transformateur de 
# colonne qui facilitera le réglage des paramètre et la recherche du transformateur (encoder) ensuite il y a le transformateur ici nous sommes sensés 
# fournir un estimateur, Nous pouvons aussi le path through le passer ou le drop laisser tomber si nous voulons. Mais puisque nous encodons, les données
# dans cet exemple, nous utilisons le OneHotEncoder. N'oubliez pas que l'estimateur que vous utiliser ici doit prendre en charge l'ajustement et la
# la transformation donc le Fit et le transforme. Ensuite il y a les colonnes [3] la liste des colonnes que nous voulons transformer.
# Dans ce cas, nous allons seulement transformer la quatrième colonne. le deuxième paramètre qui nous intéresse est le reste (remainder) il
# indique qu transformateur ce qu'il doit faire avec les autres colonnes de l0ensemble des données. Par défaut seules les colonnes qui sont
# transformées seront retourneèes par le transformateur. Toute les autres seront abondonnées. Mais nous avons la possibilité d'indiquer au 
# trasnformateur ce qu'il doit faire avec les autres colonnes. Nous pouvons soit les laisser tomber, les faire passer sans modification. Ou spécifier
# un autre estimateur si nous voulons faire plus de traitements. Comme vous pouvez le voire ici, nous avons nommé le transformateur simplement encoder
# ensuite nous vons utilisé le constructeur OneHotEncoder pour fournir une nouvelle instance comme estimateur et puis nous spécifiant que 
# seule la quatrième colonne doit être trasnsformée. Nous nous assurons également que les colonnes restantes sont passés sans modification
# une fois que nous avons construit cet objet colonne transformer nous devons ajuster et transformer l'ensemble des données. Pour ce faire nous
# avons la ligne de codes duivante :np.array(ct.fit_transform(X))
# Elle prend le jeu de données X et transforme les valeurs de l'attribut state en numérique en utilisant UnHotEncoder. De plus ce code prend toutes les
# valeurs du jeu de données X transformées et les convertit en un tableau. Après l'encodage de notre jeu de données, nous pouvons voire que la quatrième
# colonne a été transformé et qu'elle est denenue les trois premières colonnes. Les valeurs 001 représentenant New-York, les valeurs 010 éa Floride et
# 100 représente la Californie. Donc maintenant notre ensemble de données et numériques est prêt à utiliser. Nous sommes mainenant bine équipés pour
# manipuler notre jeu de données.
```

```python
# Implémentation en Python: Fractionnement des données en ensembles de formation
# La prochaine tâche importante consiste à diviser notre ensemble de données en un ensemble de formations, d'entrainements et un ensemble de tests.
# nous faisons cela afin d'entrainer notre modèle avec une partie de données appelées ensembles d'entrainement et de tester les résultats de la 
# prédiction sur l'autre ensemble de données appelées ensembles de tests.
# nous écrivons
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
# qu'avons nous fait ici, Nous avons utilisé la fonction Trains Test Split pour diviser nos données. Nous avons utilisé la fonction TranTestSplit pour 
# diviser nos données. Ici nous donnons un Test Size une valeur de 0.2, ce qui indique que 20% des données constituent l'ensemble test. 
# l'ensemble d'entraînement dans notre cas, dix données de démarrage aléatoire sont choisies comme ensemble test et 40 
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Alors qu'est qu'on a fait ici. Nous avons importé la classe Linear regression que nous allons appliquer à notre ensemble de formation, notre centre
# d'entrainement. Nous avons attribué une variable appelée régressior o à la classe linéaire régression. Nous avons ensuite utilisé la fonction
# regression fit pour ajuster l'ensemble des données d'apprentissage X train et Y Train à cette classe linéaire regression pour que le processus d'ap
# prentissage ait lieu
#                             Implémentation en Python: PREDICTION DES RESULTATS DE L'ENSEMBLE DE TEST
# Nous allons maintenant prédire les valeurs de profit de l'ensemble de tsts en utilisant le modèle formé Regressor.




```

```python
y_pred = regressor.predict(X_test)
```
